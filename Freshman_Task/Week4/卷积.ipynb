```python
k1 = np.array([
    [1,0,-1],
    [1,0,-1],
    [1,0,-1]
])
```

```python

k2 = np.array([
    [1,1,1],
    [0,0,0],
    [-1,-1,-1]
])
```

```python

def convolution(k, data,stride):
    n,m = data.shape
    img_new = []
    for i in range(0,n-3,stride):
        line = []
        for j in range(0,m-3,stride):
            a = data[i:i+3,j:j+3]
            line.append(np.sum(np.multiply(k, a)))
        img_new.append(line)
    return np.array(img_new)
```

```python
def MAXpool(size,data):
    h, w= data.shape
    new_h = h // size
    new_w = w // size
    img_new = np.zeros((new_h,new_w))
    for i in range (new_h):
        for j in range (new_w):
            img_new[i,j] = np.max(data[i*size:i*size + size,j*size:j*size + size])
    return img_new
```

```python
img_new = convolution(k1, r,2)
plt.imshow(img_new, cmap='gray')
plt.axis('off')
```

```python
img_new = MAXpool(2,img_new)
plt.imshow(img_new, cmap='gray')
plt.axis('off')
```
